**Paso 1:** Definir las variables.

```python
n_simulations = 3000  # Número de simulaciones
def f(x):
    return x**2       # Definir la función a integrar (puede ser modificada)
a = 0                 # Definir los límites de integración
b = 1
```

**Paso 2:** Formule la distribución de probabilidad de cada variables (uniforme en este caso)

**Paso 3:** Enumere la distribución acumulada de probabilidad de cada variable (no aplica en este caso)

**Paso 4:** Establezca el intervalo de números aleatorios correspondiente a cada valor de cada una de las variables.



```python
x_random = np.random.uniform(a, b)
```

**Paso 5:** Genere números aleatorios.

**Paso 6:** Simule un evento.

```python
integral_values = []
for _ in range(n_simulations):
    x_random = np.random.uniform(a, b)
    integral_values.append((b - a) * f(x_random))
```

**Paso 7:** Obtenga la gráfica de estabilización que indica que el tamaño de muestra utilizado es suficiente para garantizar la convergencia del resultado.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.arange(1, n_simulations + 1), np.cumsum(integral_values) / np.arange(1, n_simulations + 1))
plt.axhline(y=np.mean(integral_values), color='r', linestyle='--')
plt.xlabel('Número de simulaciones')
plt.ylabel('Aproximación de la integral')
plt.title('Convergencia hacia la integral')
plt.show()
```

**Paso 8:** Replique el modelo.

```python
plt.hist(integral_values, bins=np.linspace(0, 1, 31), edgecolor='black')
plt.xlabel('Aproximación de la integral')
plt.ylabel('Frecuencia')
plt.title('Histograma de Aproximaciones de la Integral')
plt.axvline(x=1/3, color='r', linestyle='--', label='Valor real de la integral')
plt.legend()
plt.show()
```

**Paso 9:** Calcule su media y desviación estandar.

```python
mean_integral = np.mean(integral_values)
std_dev_integral = np.std(integral_values)
print("Media de la integral:", mean_integral)
print("Desviación Estándar de la integral:", std_dev_integral)
```

**Paso 10:** Halle el intervalo de confianza al 95%.

```python
confidence_interval_integral = (mean_integral - 1.96 * std_dev_integral / np.sqrt(n_simulations),
                                mean_integral + 1.96 * std_dev_integral / np.sqrt(n_simulations))
print("Intervalo de confianza al 95% de la integral:", confidence_interval_integral)
```
