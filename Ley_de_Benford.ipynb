```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Función para generar números que siguen la Ley de Benford
def generate_benford_numbers(size):
    benford_numbers = []
    for _ in range(size):
        digit = np.random.choice([1,2,3,4,5,6,7,8,9], p=[0.301, 0.176, 0.125, 0.097, 0.079, 0.067, 0.058, 0.051, 0.046])
        number = digit * (10 ** np.random.uniform(0, 5))  # Genera un número con el dígito inicial correcto
        benford_numbers.append(number)
    return benford_numbers

# Generar un conjunto de datos de ejemplo
data = generate_benford_numbers(1000)

# Extraer los primeros dígitos
def get_first_digit(number):
    return int(str(number)[0])

first_digits = [get_first_digit(num) for num in data]

# Calcular las frecuencias observadas
observed_frequencies = [first_digits.count(digit) for digit in range(1, 10)]

# Frecuencias esperadas según la Ley de Benford
benford_frequencies = [math.log10(1 + 1/d) for d in range(1, 10)]
benford_frequencies = [freq * len(data) for freq in benford_frequencies]  # Escalar según el tamaño del conjunto de datos

# Graficar las frecuencias observadas vs. esperadas
plt.figure(figsize=(10, 6))
plt.bar(range(1, 10), observed_frequencies, alpha=0.7, label='Observadas')
plt.plot(range(1, 10), benford_frequencies, 'ro-', label='Esperadas (Ley de Benford)')
plt.xlabel('Primer Dígito')
plt.ylabel('Frecuencia')
plt.title('Distribución de los Primeros Dígitos según la Ley de Benford')
plt.legend()
plt.show()
```
