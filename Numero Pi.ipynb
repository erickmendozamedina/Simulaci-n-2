**6. Repita el experimento n veces.**
Así dispondrá de observaciones sobre el comportamiento del sistema, lo que permitirá entender su funcionamiento así como evaluar el desempeño del mismo frente a los diversos escenarios establecidos.

**Paso 1:** Definir las variables.

```python
n_simulations = 2000  # Número de simulaciones
n_points = 2000       # Número de puntos generados en cada simulación
```

**Paso 2:** Formule la distribución de probabilidad de cada variables (uniforme en este caso)

**Paso 3:** Enumere la distribución acumulada de probabilidad de cada variable (no aplica en este caso)

**Paso 4:** Establezca el intervalo de números aleatorios correspondiente a cada valor de cada una de las variables.

```python
x = np.random.uniform(0, 1, (n_simulations, n_points))
y = np.random.uniform(0, 1, (n_simulations, n_points))
```

**Paso 5:** Genere números aleatorios.

**Paso 6:** Simule un evento.

```python
inside_circle = (x**2 + y**2) <= 1
pi_approximations = 4 * np.sum(inside_circle, axis=1) / n_points
```

**Paso 7:** Obtenga la gráfica de estabilización que indica que el tamaño de muestra utilizado es suficiente para garantizar la convergencia del resultado.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.plot(np.arange(1, n_simulations + 1), pi_approximations.cumsum() / np.arange(1, n_simulations + 1))
plt.axhline(y=np.pi, color='r', linestyle='--')
plt.xlabel('Número de simulaciones')
plt.ylabel('Aproximación de Pi')
plt.title('Convergencia hacia Pi')
plt.show()
```

**Paso 8:** Replique el modelo.

```python
plt.hist(pi_approximations, bins=30, edgecolor='black')
plt.xlabel('Aproximación de Pi')
plt.ylabel('Frecuencia')
plt.title('Histograma de Aproximaciones de Pi')
plt.show()
```

**Paso 9:** Calcule su media y desviación estandar.

```python
mean_pi = np.mean(pi_approximations)
std_dev_pi = np.std(pi_approximations)
print("Media de Pi:", mean_pi)
print("Desviación Estándar de Pi:", std_dev_pi)
```

**Paso 10:** Halle el intervalo de confianza al 95%.

```python
confidence_interval = (mean_pi - 1.96 * std_dev_pi / np.sqrt(n_simulations),
                       mean_pi + 1.96 * std_dev_pi / np.sqrt(n_simulations))
print("Intervalo de confianza al 95%:", confidence_interval)
```
